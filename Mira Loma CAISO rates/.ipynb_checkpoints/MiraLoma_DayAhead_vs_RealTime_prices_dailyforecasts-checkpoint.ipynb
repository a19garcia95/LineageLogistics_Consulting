```python
#define the class to make the 'get' request
#This code uses the documentation specific in "Interface Specification for OASIS" provided by CAISO
class CaisoRateWorker:

    def __init__(self):
        self.t_now = utc.localize(datetime.now())
        pass
    
    def get_historical_zips(self,t0,tf,market):
        assert tf < self.t_now, "tf is in the future. Use 'group_future_queries'"
        assert market == 'Day_Ahead'or market == '5min_Real_Time' or market == '15min_Real_Time'
        # create a new directory to put our results in
        path = ('./' + str(t0.year) + '/oasis_results/' + market + '/')
        if not os.path.exists(path):
            os.makedirs(path)
        #break the API call into 15 day chunks for zip file storage
        t_list = self.chunkify(t0,tf)
        for times in t_list:
            if market == 'Day_Ahead':
                fname = (times[0].strftime("%Y%m%d") + "_" + times[1].strftime("%Y%m%d") + "_" + 'PRC_LMP_DAM')
            elif market == '5min_Real_Time':
                fname = (times[0].strftime("%Y%m%d") + "_" + times[1].strftime("%Y%m%d") + "_" + 'PRC_INTVL_LMP')
            elif market == '15min_Real_Time':
                fname = (times[0].strftime("%Y%m%d") + "_" + times[1].strftime("%Y%m%d") + "_" + 'PRC_RTPD_LMP')
            
            #check if we have the zip file already
            ##always get the period which we are in right now##
            exists = glob((path + fname + '*'))                
                
            if len(exists)>0 and times[1] < self.t_now:
                print('Already have the zip file for ' + fname)
                continue
            elif len(exists)>0 and times[1] >= self.t_now:
                print('Removing the files for the current time period')
                for file in exists: os.remove(file)
                start_time = times[0].strftime("%Y%m%dT%H:%M%z")
                end_time = times[1].strftime("%Y%m%dT%H:%M%z")
                print("grabbing " + fname)
                request = self.get_data(start_time,end_time,market)
                self.save_zip(request,path)                    
            else:
                start_time = times[0].strftime("%Y%m%dT%H:%M%z")
                end_time = times[1].strftime("%Y%m%dT%H:%M%z")
                print("grabbing " + fname)
                request = self.get_data(start_time,end_time,market)
                self.save_zip(request,path)
                
        return path
            
    def get_data(self,start_time,end_time,market):
        try:
            ('Day_Ahead','5min_Real_Time','15min_Real_Time',).index(market)
        except ValueError:
            print('Only valid options are Day_Ahead and Real_Time')
        if market == 'Day_Ahead':
            self.query_name = 'PRC_LMP'
            self.market_id = 'DAM'
        elif market == '5min_Real_Time':
            self.query_name = 'PRC_INTVL_LMP'
            self.market_id = 'RTM'
        elif market == '15min_Real_Time':
            self.query_name = 'PRC_RTPD_LMP'
            self.market_id = 'RTPD'
        
        query = ("http://oasis.caiso.com/oasisapi/SingleZip?" +
                         "queryname=" + 
                          self.query_name + "&" + 
                         "startdatetime={}&" +
                         "enddatetime={}&" + 
                         "market_run_id=" + 
                          self.market_id + "&" + 
                         "version=1&" + 
                         "node=DLAP_SCE-APND")
        
        request = requests.get(query.format(start_time, end_time), stream=True)
        #need to wait 5 seconds between queries
        time.sleep(5)
        print('status code: ', request.status_code)
        if request.status_code != 200:
            print(request.text)
        assert request.status_code == 200, request.status_code #429 ERROR CODE TOO MAY API REQUEST
        return request
    
    def save_zip(self, request, path):
        # create directory for zip files if it doesn't exist
        if not os.path.exists(path):
            os.makedirs(path)
            
        #makes sure we have the raw data we need
        dat = io.BytesIO(request.content)
        assert zipfile.is_zipfile(dat) == True
        z = zipfile.ZipFile(dat)
        fname = z.namelist()[0]
        z.extractall(path)
        print("Saved " + fname)

    def organize_all(self,path, market):
        periods = {}
        for file in glob((path + "*.xml")):
            with open(file) as f:
                contents = f.read()
                xml_dict = xmltodict.parse(contents)
                
                for report in xml_dict['OASISReport']['MessagePayload']['RTO']['REPORT_ITEM']:
                    #<class 'collections.OrderedDict'> shows up and bugs the code??
                    if isinstance(report['REPORT_DATA'], list) == False:
                        continue
                            
                    for data in report['REPORT_DATA']:
                        if data['DATA_ITEM'] != 'LMP_PRC':
                            continue
                        
                        #This API returns timezones in UTC
                        key = datetime.strptime(data['INTERVAL_START_GMT'][:16],"%Y-%m-%dT%H:%M")
                        if key not in periods:
                            periods[key] = {
                            'rate_start': data['INTERVAL_START_GMT'][:16],
                            'rate_end': data['INTERVAL_END_GMT'][:16],
                            'rate': data['VALUE']
                        }
        return periods
    
    def to_dataframe(self,data):
        df = pd.DataFrame.from_dict(data,orient="index")
        #just use rate start as index and convert rates to floats
        df_f = pd.to_numeric(df['rate'],errors='ignore')
        return df_f
    
    def chunkify(self,t0,tf):
        #only deal with data in the same year for now
        assert t0.year == tf.year, "let's do this year by year"
        t_list = []
        
        #always start at the beginning of the month
        t0_get = t0.replace(day=1,hour=0)
        #for data within a month
        if tf.month == t0.month and tf.month == 12:
            tf_get = tf.replace(year = tf.year + 1, month = 1, day=1,hour=0)
        
        elif tf.month == t0.month and tf.month != 12:
            tf_get = tf.replace(month = tf.month + 1, day=1,hour=0)
        #for data in different months
        else:
            if tf.month != 12:
                tf_get = tf.replace(month = tf.month + 1, day=1,hour=0)
            else:
                tf_get = tf.replace(year = tf.year + 1, month = 1, day=1,hour=0)
        
        #break the API call into 15 day chunks
        val = pd.date_range(start=t0_get, end=tf_get, freq=pd.offsets.SemiMonthBegin())
        
        for i in range(len(val)-1):
            t_list.append((val[i] ,val[i+1]))
        
        return t_list
```

```python
#### Initial time series checks ####
#### Is the time series stationary? 
#### Note: it looks relatively stationary with
####    the daily data that I have right now.
####    More history might shed more light on this trend.

def test_stationarity(timeseries,window_size=7):
    
    #Determing rolling statistics
    rolmean = timeseries.rolling(window=window_size).mean()
    rolstd = timeseries.rolling(window=window_size).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
    #Perform Dickey-Fuller test:
    print 'Results of Dickey-Fuller Test:'
    dftest = adfuller(timeseries, autolag='AIC')
    dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
    for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print dfoutput
```

```python
## Try to address the nonstationary nature of the DAM ts
DAM_diff1 = summary_by_date['cost_DAM'] - summary_by_date['cost_DAM'].shift(7)
DAM_diff1.dropna(inplace=True)
test_stationarity(DAM_diff1,7)

###Conclusion- Taking the difference with a week lag 
###seems to look much better will use this for forecasts
```

```python
def search_first(list,value_to_beat):
    sub_list = [index for index, item in enumerate(list) if item > value_to_beat and index > 1]
    if len(sub_list)>0:
        first = sub_list[0]
        return first
    else:
        return 'not found'

def correlation_funcs(timeseries):
    ###autocorrelation and partial autocorrelation functions for DAM###
    acfs = acf(timeseries, nlags=20)
    pcfs = pacf(timeseries, nlags=20)
    
    value_to_beat = 1.96/np.sqrt(len(timeseries))
    acf_lag_choice = search_first(acfs,value_to_beat)
    pacf_lag_choice = search_first(pcfs,value_to_beat)
    
    #Plot ACF: 
    plt.subplot(121) 
    plt.plot(acfs)
    plt.axhline(y=0,linestyle='--',color='gray')
    plt.axhline(y=-1.96/np.sqrt(len(timeseries)),linestyle='--',color='gray')
    plt.axhline(y=1.96/np.sqrt(len(timeseries)),linestyle='--',color='gray')
    plt.title('Autocorrelation Function')
    
    #Plot PACF:
    plt.subplot(122)
    plt.plot(pcfs)
    plt.axhline(y=0,linestyle='--',color='gray')
    plt.axhline(y=-1.96/np.sqrt(len(timeseries)),linestyle='--',color='gray')
    plt.axhline(y=1.96/np.sqrt(len(timeseries)),linestyle='--',color='gray')
    plt.title('Partial Autocorrelation Function')
    plt.tight_layout()
    
    return(pacf_lag_choice,acf_lag_choice)
```

```python
####DAM ANALYSIS####

###Choose p and q
p,q=correlation_funcs(DAM_diff1)
### p  The lag value where the PACF chart crosses the upper confidence interval for the first time.
### q - The lag value where the ACF chart crosses the upper confidence interval for the first time.
```

```python
## DAM ANALYSIS CONT'd ##
model = ARIMA(DAM_diff1, order=(p, 1, q))  
results_DAM = model.fit(disp=-1)  
RSS = (results_DAM.fittedvalues-DAM_diff1)**2
RSS_sum = sum(RSS.dropna())
plt.plot(DAM_diff1)
plt.plot(results_DAM.fittedvalues, color='red')
plt.title('RSS: %.4f'% RSS_sum)

###Now get back to predicted cost###
predictions_ARIMA_diff = pd.Series(results_DAM.fittedvalues, copy=True)
predictions_ARIMA = predictions_ARIMA_diff + summary_by_date['cost_DAM'].shift(7)
forecast_ARIMA_7days = results_DAM.forecast(7)[0]
future_DAM_predicted_cost = forecast_ARIMA_7days + summary_by_date['cost_DAM'][-7:]
future_DAM_predicted_cost.index = future_DAM_predicted_cost.index+timedelta(days=7)
###other output of forecast results_DAM.forecast()[1] & [2] are standard error and confidence
```
